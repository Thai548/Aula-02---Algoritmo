***Colhimento de dados abaixo:***

```python
#Colher os dados:

nota1 = float(input('1º Nota: '))
nota2 = float(input('2º Nota: '))
nota3 = float(input('3º Nota: '))
nota4 = float(input('4º Nota: '))
```

```python
# Cálcular a média

media= (nota1+nota2+nota3+nota4) /4

print(f'Média: {media} ')
```

```python
#Condição de aprovado ou não.

print(f'Resultado Final: ')

if media >=7:
  print('Aprovado')
elif 5 <= media < 7:
  print('Recuperação')
else:
  print('Reprovado :/ ')

```
